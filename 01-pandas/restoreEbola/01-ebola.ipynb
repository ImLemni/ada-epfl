The `DATA_FOLDER/ebola` folder contains summarized reports of Ebola cases from three countries (Guinea, Liberia and Sierra Leone) during the recent outbreak of the disease in West Africa. For each country, there are daily reports that contain various information about the outbreak in several cities in each country.

Make sure you handle all the different expressions for *new cases* and *deaths* that are used in the reports.

```python
import os
import pandas as pd
import numpy as n
from typing import Generator

def get_file_paths(root: str) -> Generator[str, None, None]:
    """
    Returns the paths for all the files in the subtree of `root`
    """
    for (dir_path, _, file_names) in os.walk(root):
        yield from [os.path.join(dir_path, file_name) for file_name in file_names]

gn_file_paths = [*get_file_paths("./data/guinea_data")]
lr_file_paths = [*get_file_paths("./data/liberia_data")]
sl_file_paths = [*get_file_paths("./data/sl_data")]
```

```python
DATE: str = "date"
"""Name of the column used for the date"""

VARIABLE: str = "variable"
"""Name of the input column used for the variable"""

COUNTRY: str = "country"
"""Name of the output column used for the country"""

REGION: str = "region"
"""Name of the output column used for the region"""

def get_or_create_row(frame: pd.core.frame.DataFrame, region, date):
#     result = frame.query("region = @region")
    print(len(frame.query("region == @region")))
    new_row = pd.DataFrame([{COUNTRY: "sl", REGION: region, DATE: date}])
    frame = pd.concat([frame, new_row])
    return frame, new_row

def normalize_sl(frame: pd.core.frame.DataFrame):
    result = pd.DataFrame()
    result[COUNTRY] = None
    result[REGION] = None
    result[DATE] = None

    for index, row in frame.iterrows():
        date = row[DATE]
        var_name = row[VARIABLE]
        if var_name not in result:
            result[var_name] = None
        
        for col_name, value in row.iteritems():
            if col_name in {DATE, VARIABLE}:
                continue
            result, updated_row = get_or_create_row(result, col_name, date)
            updated_row[var_name] = value
#             result = pd.concat([result, updated_row])
            break
    return result

sl0 = pd.read_csv(sl_file_paths[0])
sl1 = normalize_sl(sl0)
# print(sl0)
sl1
```
